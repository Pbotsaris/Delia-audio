# **IIR Filters**

## **Canonical Filters**

<img src="diagrams/canonical_filters.png" width="70%">

Canonical filters are **IIR (Infinite Impulse Response) systems**, and they represent one of the simplest types of filters to implement. The design involves two main steps: calculating the **feedback portion** first and then calculating the **feedforward portion**.

**Step 1: Feedback (Recursive) Calculation**
First, we calculate the feedback portion to generate an intermediate signal  $x_h(n)$. This is done recursively:

<br><br>
$
\large
x_h(n) = x(n) - a_1 x_h(n - 1) - a_2 x_h(n - 2)
\large
$

<br><br>

Here,  $a_1 $ and  $a_2 $ are the **feedback coefficients**. This equation ensures that the current value of  $x_h(n) $ depends not only on the current input  $x(n) $ but also on the past values of the intermediate signal, introducing a **memory effect** in the system.

**Step 2: Feedforward (Non-Recursive) Calculation**
Once the feedback portion is computed, we calculate the final output  $y(n) $ using the intermediate signal  $x_h(n) $:

<br><br>
$
\large
y(n) = b_0 x_h(n) + b_1 x_h(n - 1) + b_2 x_h(n - 2)
\large
$
<br><br>

Here,  $b_0 $,  $b_1 $, and  $b_2 $ are the **feedforward coefficients**. This equation combines the current and past values of  $x_h(n) $ to produce the final output signal  $y(n) $.

These leads to the following transfer function:
<br> <br>

$$
\large
H(z) = \frac{b_0 + b_1 z^{-1} + b_2 z^{-2}}{1 + a_1 z^{-1} + a_2 z^{-2}}
\large
$$

<br> <br>

### **The $K$ Coefficient**

For **HighPass** & **Lowpass** filters, the $K$ coefficient is directly related to the the cut off frequency of the a canonical filter. Specifically, it represents a **scaling factor** that adjusts the relative weight of the current and past values in the filter, and it is usually calculated from the filter's cutoff frequency.

<br> <br>
$$
\large
K = \tan(\pi f_c/ f_s)
\large
$$

<br> <br>

where $fc$ is the cut off frequency and $fs$ the sampling frequency.

For **Allpass** filters, $K$ is related to adjusting the phase of signal rather than the amplitude. It controls the frequency $fc$ when $-90^\circ$ phase shift is reached.

Using $K$, we adjust the $a$ (feedback coefficients) and $b$ (feedforward coefficients) to achieve the desired cutoff effect in the canonical filter. Below is a table with the coefficient values:
 


| Filter     | b0           |  b1        | a1             |
|------------|--------------|------------|----------------|
| Lowpass    | $K/(K+1)$    | $K/(K+1)$  | $(K-1)/(K+1)$  | 
| Highpass   | $1/(K+1)$    |$-1/(K+1)$ | $(K-1)/(K+1)$  |
| Allpass    | $(K-1)/(K+1)$|     1      | $(K-1)/(K+1)$  |

### **Second-order Filters and the $Q$ Factor**



For **second-order filters**, in addition to the cutoff frequency or center frequency, we need a parameter $Q$, which has slightly different meanings depending on the filter type:

- **Lowpass** and **Highpass**: $Q$  controls the height of the resonance. For $Q = \frac{1}{\sqrt{2}}$ , the filter is maximally flat up to the cutoff frequency. Lower values of $Q$  result in higher pass-band attenuation, while higher values lead to amplification around the cutoff frequency $f_c$ , introducing more resonance.

- **Bandpass** and **Bandreject**: In these filters, $Q$  is related to the bandwidth $f_b$  and is defined as $Q = \frac{f_c}{f_b}$ , where $f_c$  is the center frequency and $f_b$  is the bandwidth. In other words, $Q$  is the inverse of the relative bandwidth $\frac{f_b}{f_c}$ . A higher $Q$  results in a narrower bandwidth, while a lower $Q$  broadens the bandwidth.

- **Allpass**: In an allpass filter, $Q$  controls the bandwidth, but in this case, itâ€™s associated with the phase shift. It defines the points where a $\pm 90^\circ$  phase shift occurs relative to the $-180^\circ$  phase shift at $f_c$ .


While cannonical filters are simple, the calculation of their coeeficients from paramets like cutt-off and reasonance is not. The table below presents the calculation to to paramaterize $K$ and $Q$.


| Filter   | $b_0$                                 | $b_1$                                 | $b_2$                                 | $a_1$                                | $a_2$                                |
|----------|---------------------------------------|---------------------------------------|---------------------------------------|--------------------------------------|--------------------------------------|
| Lowpass  | $\frac{K^2 Q}{K^2 + Q + K + Q}$       | $\frac{2K^2 Q}{K^2 + Q + K + Q}$      | $\frac{K^2 Q}{K^2 + Q + K + Q}$       | $\frac{2Q(K^2 - 1)}{K^2 + Q + K + Q}$ | $\frac{K^2 Q - K + Q}{K^2 Q + K + Q}$ |
| Highpass | $\frac{Q}{K^2 + Q + K + Q}$       | $\frac{2Q}{K^2 + Q + K + Q}$      | $\frac{Q}{K^2 + Q + K + Q}$       | $\frac{2Q(K^2 - 1)}{K^2 + Q + K + Q}$ | $\frac{K^2 Q - K + Q}{K^2 Q + K + Q}$ |
| Bandpass | $\frac{K}{K^2 + Q + K + Q}$       | $0$      | $-\frac{K}{K^2 + Q + K + Q}$       | $\frac{2Q(K^2 - 1)}{K^2 + Q + K + Q}$ | $\frac{K^2 Q - K + Q}{K^2 Q + K + Q}$ |
| Bandreject| $\frac{Q(1 + K^2)}{K^2 + Q + K + Q}$       | $\frac{2Q(K^2 - 1)}{K^2 + Q + K + Q}$      | $\frac{Q(1 + K^2)}{K^2 + Q + K + Q}$       | $\frac{2Q(K^2 - 1)}{K^2 + Q + K + Q}$ | $\frac{K^2 Q - K + Q}{K^2 Q + K + Q}$ |  
| Allpass| $\frac{K^2Q- K + Q}{K^2 + Q + K + Q}$       | $\frac{2Q(K^2 - 1)}{K^2 + Q + K + Q}$      | $1$       | $\frac{2Q(K^2 - 1)}{K^2 + Q + K + Q}$ | $\frac{K^2 Q - K + Q}{K^2 Q + K + Q}$ |  


The following filters presented in this notebook are more complex, but they offer a more flexible and intuitive way to parameterize their behavior.


```python
import numpy as np
import librosa
import matplotlib.pyplot as plt
from IPython.display import Audio, display
from scipy.signal import freqz

samples, sr = librosa.load('audio_examples/voice.wav', sr=44100)
samples = samples[: 2 * sr] # two seconds only

cutoff = 80 # hz
Q =  0.6 # flatten is 1/sqrt(2) ~= 0.707

# K = tan(Ï€*fc/fs)
K = np.tan(np.pi * cutoff / sr)

# 2nd order
def lowpass_coefs(K, Q):
    den = K**2 * Q + K + Q
    
    a1 = (2 * Q * (K**2 - 1)) / den
    a2 = (K**2 * Q - K + Q) / den
    
    b0 = (K**2 * Q) / den
    b1 = (2 * K**2 * Q) / den
    b2 = (K**2 * Q) / den

    return a1, a2, b0, b1, b2

def cannonical_lowpass(samples, K, Q):
    a1, a2, b0, b1, b2 = lowpass_coefs(K, Q)
    N = len(samples)

    x_h = np.zeros(N)
    y = np.zeros(N)
     
    for n in range(N):
        if n == 0:
            x_h[n] = samples[n]
            y[n] = b0 * x_h[n]
        elif n == 1:
            x_h[n] = samples[n] - a1 * x_h[n - 1]
            y[n] = b0 * x_h[n] + b2 * x_h[n - 1]
        else:
            x_h[n] = samples[n] - a1 * x_h[n - 1] - a2 * x_h[n - 2]
            y[n] = b0 * x_h[n] + b1 * x_h[n - 1] + b2 * x_h[n - 2]

    return y

def samples_to_time(samples):
    times = []

    for s in range(len(samples)):
        times.append(s / sr)
        
    return times

def plot_frequency_response(K, Q):
    a1, a2, b0, b1, b2 = lowpass_coefs(K, Q)
    b = [b0, b1, b2]
    a = [1, a1, a2]  # first coefficient of `a` is always 1 for normalized IIR filters
    w, h = freqz(b, a, worN=8000) # 8000 evenly spaced points between 0 and Ï€ radians/sample
    plt.plot(0.5 * sr * w / np.pi, np.abs(h), 'b')
    plt.title("Frequency Response of the Filter")
    plt.xlabel("Frequency (Hz)")
    plt.xscale('log')  
    plt.xticks([10, 100, 1000, 10000], [10, 100, 1000, 10000])  # Custom ticks in Hz
    
    plt.ylabel("Gain")
    plt.grid()

filtered = cannonical_lowpass(samples, K, Q)

plt.figure(figsize=(8, 4))
plot_frequency_response(K, Q)

plt.figure(figsize=(8,4))
plt.plot(samples_to_time(samples), samples)
plt.title("Original Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")


plt.figure(figsize=(8,4))
plt.plot(samples_to_time(samples), filtered)
plt.title("Filtered Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")

plt.show()

print("Original")
display(Audio(samples, rate=sr))
print("Filtered")
display(Audio(filtered, rate=sr))
```

## **State Variable Filter**


<img src="diagrams/state_variable_filter.png" width="70%">

An attractive alternative to canonical filters is the **state variable filter**, which combines second-order lowpass, highpass, and bandpass outputs for the same $f_c$ and $Q$. The difference equations for these filter modes are:

**Lowpass** ($y_l$):

$
\large
y_l(n) = F_1 \cdot y_b(n) + y_l(n - 1)
\large
$
<br><br>

**Bandpass** ($y_b$):

$
\large
y_b(n) = F_1 \cdot y_h(n) + y_b(n - 1)
\large
$
<br><br>

**Highpass** ($y_h$):

$
\large
y_h(n) = x(n) - y_l(n - 1) - Q_1 \cdot y_b(n - 1)
\large
$
<br><br>

Here, $Q_1$ and $F_1$ are coefficients related to the tuning parameters $f_c$ (cutoff) and $Q$ (resonance) by:

$$
\large
\begin{aligned}
    F_1 &= 2 \sin\left(\frac{\pi f_c}{f_s}\right) \quad &Q_1 &= \frac{1}{Q}
\end{aligned}
\large
$$


The **transfer functions** for these filter modes are as follows:

**Lowpass** ($H_l(z)$):

$
\large H_l(z)\large \LARGE = \frac{r^2}{1 + (r^2 - q - 1) \cdot z^{-1} + qz^{-2}}\LARGE
$
<br><br>

**Bandpass** ($H_b(z)$):

$
\large H_b(z)\large \LARGE = \frac{r \cdot (1 - z^{-1})}{1 + (r^2 -  q - 1) \cdot z^{-1} + qz^{-2}}\LARGE
$
<br><br>

**Highpass** ($H_h(z)$):

$
\large H_h(z)\large \LARGE = \frac{(1 - z^{-1})^2}{1 + (r^2 - q - 1) \cdot z^{-1} + qz^{-2}}\LARGE
$
<br><br>

with

$
\large
\begin{aligned}
    r &= F_1 \quad &q &= 1 - F_1 Q_1
\end{aligned}
\large
$

This structure is effective not only in the filtering process but also because of the simple relation between control parameters and tuning coefficients.

#### **Stability Considerations**

At higher cutoff frequencies and smaller $Q$ factors, the filter becomes unstable. The *usability limit* is given by:

<br><br>
$$
\large
F_1 < 2 - Q_1
\large
$$
<br><br>

For example, with a cutoff frequency $f_c=1000$ Hz, a sampling rate of $f_s=44100$ Hz, and $Q=0.4$, the tuning parameters are:

<br> <br>
$$
\large
\begin{aligned}
    F_1 &= 2 \sin\left(\frac{\pi \cdot 1000}{44100}\right) \approx 0.142355 \quad &Q_1 &= \frac{1}{0.4} = 2.5
\end{aligned}
\large
$$
<br> <br>

As you can see, this limitation is generally not a problem because the tuning frequency ($F_1$) is usually small at commonly used sample rates ($f_s$), and the $Q$ factor is typically set to sufficiently high values.

```python
fs = sr
fc = 1000 # hz
Q = 0.5

F1 = 2 * np.sin(np.pi * fc / fs)
Q1 = 1 / Q
epsilon = 1e-40  # avoid division by zero


def state_variable_filter(x, F1, Q1):
    N = len(samples)
    y_h = np.zeros(N)
    y_l = np.zeros(N)
    y_b = np.zeros(N)

    for n in range(N):
        if n == 0:
            y_h[n] = x[n] # y_l[n - 1] and y_b[n-1] = 0
            y_b[n] = F1 * y_h[n]
            y_l[n] = F1 * y_b[n]
            continue

        y_h[n] =  x[n] - y_l[n - 1] - Q1 * y_b[n - 1]
        y_b[n] =  F1 * y_h[n] + y_b[n - 1]
        y_l[n] = F1 * y_b[n] + y_l[n - 1]

    return y_h, y_b, y_l


highpass_out, bandpass_out, lowpass_out = state_variable_filter(samples, F1, Q1) 

# manually calculating the frequency response (instead of scipy.freqz)
def freq_response(F1, Q1, fs, n_points=1024):
    w = np.linspace(0, np.pi, n_points) # radian / sample

    # Outs
    H_l = np.zeros_like(w, dtype=complex)
    H_b =  np.zeros_like(w, dtype=complex)
    H_h = np.zeros_like(w, dtype=complex)

    
    q = 1 - F1 * Q1  
    r = F1

    for n, omega in enumerate(w): 
        z = np.exp(1j * omega)  # z = e^(jÏ‰)

        den = (1 + (r**2 - q - 1) * z**-1 + q * z**-2 + epsilon)

        # Transfer functions
        H_l[n] = r**2 / den  
        H_b[n] = (r * (1 - z**-1)) / den  
        H_h[n] = ((1 - z**-1)**2) / den  

    freq_hz = (w * fs)/ (2 * np.pi)
    return freq_hz, H_h, H_b, H_l

freqs, response_high, response_band, response_low = freq_response(F1, Q1, fs) 

plt.figure(figsize=((8,4)))
plt.plot(freqs, np.abs(response_high), label ="Highpass")
plt.plot(freqs, np.abs(response_band), label="Bandpass")
plt.plot(freqs, np.abs(response_low), label="Lowpass")
plt.xscale('log')  
plt.xticks([10, 100, 1000, 10000, 20000], [10, 100, 1000, 10000, 20000])  # Custom ticks in Hz
plt.legend()

plt.title("State Variable Filter: Freq Response for fc = 1Khz and Q = 0.4")
plt.xlabel("Frequency(Hz)")
plt.ylabel("Magnitude")
plt.grid(True)
plt.show()


print("Original")
display(Audio(samples, rate=sr))

print("Highpass Output")
display(Audio(highpass_out, rate=sr))

print("Bandpass Output")
display(Audio(bandpass_out, rate=sr))

print("Lowpass Output")
display(Audio(lowpass_out, rate=sr))
```

## **Normalization**

Filters are generally designed in the frequency domain, where their primary purpose is to shape the frequency content of the signal. However, a crucial practical consideration is the effect of the filter on the signal's loudness. Without accounting for this, the filtered signal may become too quiet or too loud, making the result less usable, even if the desired frequency effect is achieved.

To address this, **normalization** is applied, which compensates for any amplitude changes introduced by the filter. This is done by scaling the filter so that its norm, often  $L_p$, is equal to 1:

$$
\LARGE
L_p = \left( \frac{1}{2\pi} \int_{-\pi}^{\pi} |H(e^{j\omega})|^p \, d\omega \right)^{\frac{1}{p}} = 1
\LARGE
$$

Here, the most common choices for  $p $ are:
-  $L_2 $: The **energy norm**, which considers the overall energy in the frequency response.
-  $L_\infty $: The **maximum norm**, which looks at the peak gain of the filter:  $L_\infty = \max(|H(e^{j\omega})|) $.

To normalize signal loudness, the $L_2$ norm is commonly used, ideal for broadband signals and most music applications, as it keeps the overall energy consistent.

The $L_\infty$ norm focuses on normalizing the **peak** of the frequency response, preventing overloading by limiting excessive amplification at any frequency.

**Note**: Applying the $L_2$ norm directly like this is not the most practical approach in DSP implementations. A more practical approach, using first-order low-pass filters and amplitude corrections for normalization, is described below.


## **$L_2$ Normalization of a State Variable Filter**

<img src="diagrams/normalization.png" width="60%">

The normalization of the **state variable filter (SVF)** involves two key steps to ensure stability, especially at extreme settings:

1. **First-Order Low-Pass Filter for  $f_c$  Normalization**:
    - A **first-order low-pass filter** is applied to the input signal to stabilize the filter's response at the cutoff frequency  $f_c$. This step ensures predictable behavior, particularly when the damping factor is very low.
<br><br>
2. **Amplitude Correction Using  $\sqrt{\zeta}$**:
   - To prevent excessive gain at  $f_c$ , the filterâ€™s output is scaled by  $\sqrt{\zeta}$ . This correction ensures stability and prevents amplification when the filter operates with very low damping values, where gains can reach up to **74 dB**.


This normalization scheme allows the filter to operate reliably even with damping factors as low as  $10^{-4}$ , maintaining stable performance and preventing distortion at high gains.

**NOTE**: Still need to get a working example going here and how to properly apply the normalization to the frequency respoinse.

```python
# Parameters
fs = 48000
fc = 1000
zeta = 0.5
F1 = 2 * np.sin(np.pi * fc / fs)
Q1 = 1 / zeta

# Frequency response without normalization
freqs, H_h_unnorm, H_b_unnorm, H_l_unnorm = freq_response(F1, Q1, fs)

# Applying normalization: low-pass filter + amplitude correction
sqrt_zeta = np.sqrt(zeta)
H_h_norm = H_h_unnorm * np.sqrt(zeta)
H_b_norm = H_b_unnorm * np.sqrt(zeta)
H_l_norm = H_l_unnorm * np.sqrt(zeta)

# Plot the unnormalized and normalized responses
plt.figure(figsize=(10, 6))
plt.plot(freqs, np.abs(H_h_unnorm), label="Unnormalized Highpass", linestyle='--')
plt.plot(freqs, np.abs(H_h_norm), label="Normalized Highpass")
plt.plot(freqs, np.abs(H_b_unnorm), label="Unnormalized Bandpass", linestyle='--')
plt.plot(freqs, np.abs(H_b_norm), label="Normalized Bandpass")
plt.plot(freqs, np.abs(H_l_unnorm), label="Unnormalized Lowpass", linestyle='--')
plt.plot(freqs, np.abs(H_l_norm), label="Normalized Lowpass")
plt.xscale('log')
plt.legend()
plt.title("Unnormalized vs Normalized Frequency Response")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.grid(True)
plt.show()
```

### **Allpass-Based Filters**:
Allpass-based filters are type of **parametric filter** that allow easy tuning of **cutoff frequency** and **bandwidth** by adjusting one or two coefficients. The **signal flow graph** directly controls the cutoff frequency and bandwidth of the corresponding filter, making them efficient and ideal for real-time applications with minimal computational complexity.

**Key Points**:
- **Parametric filter structures** use a simple signal flow where a coefficient directly controls filter behavior (cutoff and bandwidth).
- **Real-time control** is possible by adjusting just a few coefficients, without needing complex recalculations.
- **Allpass filters** (first-order or second-order) form the basis. They modify the **phase** of the signal, while leaving the **amplitude unchanged**.
- When combined with a **direct signal path**, these allpass filters can create **lowpass, highpass, bandpass**, and **band-reject filters** that are easily tunable and efficient.

#### **First-Order Allpass**

<img src="diagrams/allpass-first.png" width="70%">

In an allpass filter, the input signal's **amplitude** remains unchanged, but its **phase** is modified, meaning parts of the signal are delayed or advanced depending on the frequency.

- **Low Frequencies**: Smaller phase shift (closer to $0^\circ$)
- **High Frequencies**: Larger phase shifts (closer to $-180^\circ$)

A first-order allpass filter is given by the following **difference equation** (similar to the canonical filters):

$$
\large
x_h(n) = x(n) - c \cdot x_h(n - 1)
\large
$$

$$
\large
y(n) = c \cdot x_h(n) + x_h(n - 1)
\large
$$

and the corresponding **transfer function**:

$$
\large
A(z) = \frac{z^{-1} + c}{1 + c \cdot z^{-1}}
\large
$$

Where the **coefficient $c$:**

$$
\large
c = \frac{1 - \tan(\pi \cdot f_c /f_s)}{1 + \tan(\pi \cdot f_c/f_s)}
\large
$$

The coefficient $ c $ controls the **phase response** of the filter, specifically determining the frequency where the phase shift passes through $ -90^\circ $.

- **Small $ c $** values (closer to 0): The transition to $ -90^\circ $ occurs at **lower frequencies**, affecting the lower end of the spectrum more.
- **Larger $ c $** values (closer to 1): The phase shift is delayed, causing $ -90^\circ $ to occur at **higher frequencies**, affecting the higher end of the spectrum.


```python
fs = 44100
# in the middle
fc = fs * 0.1

n_samples = 100

tan = np.tan(np.pi * fc / fs)
c = (tan - 1) / (tan + 1)

def allpass_first_order(x, c, fs):
    N = len(x)
    x_h = np.zeros(N)
    y = np.zeros(N)

    for n in range(1, N):
        x_h[n] = x[n] - c * x_h[n - 1]
        y[n] = c * x_h[n] + x_h[ n- 1]

    return y

def allpass_freq_response(c, fs, n_points=1024):
    w = np.linspace(0, np.pi, n_points)
    A = np.zeros_like(w, dtype=complex)

    for n, omega in enumerate(w):
        z = np.exp(1j * omega)
        A[n] = (z**-1 + c) / (1 + c * z**-1)

    freqs = (w * fs)/ (2 * np.pi)
    return freqs, w, A

freqs, w, A = allpass_freq_response(c, fs)
phase_radians = np.angle(A)
phase_degrees = np.degrees(phase_radians)
group_delay = -np.diff(phase_radians) / np.diff(w)

# filtering just the first 100 samples for plotting
filtered = allpass_first_order(samples[n_samples:n_samples + n_samples], c, fs)

plt.figure(figsize=(8, 8))

plt.subplot(5, 1, 1)
plt.title("Allpass: Magnitude Response")
plt.plot(freqs, np.abs(A))
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.xlim(0,20000)
plt.grid(True)

plt.subplot(5, 1, 2)
plt.title("Allpass: Phase Response")
plt.plot(freqs, phase_degrees)
plt.xlabel("Frequency(Hz)")
plt.ylabel("Phase (Degrees)")
plt.xlim(0, 20000)
plt.grid(True)

plt.subplot(5, 1, 3)
plt.title("Allpass: Group Delay")
plt.plot(freqs[:-1], group_delay)
plt.xlabel("Frequency(Hz)")
plt.ylabel("Group Delay (Samples)")
plt.xlim(0, 20000)
plt.grid(True)

plt.subplot(5, 1, 4)
plt.title("Original: 100 samples")
plt.plot(np.linspace(0, n_samples, n_samples), samples[n_samples:n_samples + n_samples])
plt.xlabel("Samples")
plt.ylabel("Amplitude")

plt.subplot(5, 1, 5)
plt.title("Filtered: 100 samples")
plt.plot(np.linspace(0, n_samples, n_samples), filtered[:n_samples])
plt.xlabel("Samples")
plt.ylabel("Amplitude")

plt.tight_layout()
```

### **First-Order Allpass-based Filter: Highpass & Lowpass**

<img src="diagrams/allpass-based-first-order.png" width="50%">

A first-order lowpass or highpass filter can be achieved by adding or subtracting the output signal from the input signal of a first-order allpass filter. The allpass filter progressively shifts the phase of frequencies, with higher frequencies experiencing more phase shift.

- **Lowpass filter**: When the output of the allpass filter is **added** to the input, the higher frequencies, which have undergone significant phase shifts, experience **phase cancellation**, attenuating the high end and allowing the lower frequencies to pass.

- **Highpass filter**: When the output of the allpass filter is **subtracted** from the input, the lower frequencies, which have less phase shift, are canceled out, letting the higher frequencies pass.

The filtering effect occurs due to **phase cancellation** between the input and the phase-shifted output at different frequencies.

The **Transfer Function** for a first-order lowpass or highpass filter based on an allpass filter is given by:

$$
\large
H(z) = \frac{1}{2} \left(1 \pm A(z)\right)
\large
$$

For a **lowpass filter (LPF)**, we use the **addition**:

$H_{\text{LPF}}(z) = \frac{1}{2} \left(1 + A(z)\right) $

For a **highpass filter (HPF)**, we use the **subtraction**: 

$H_{\text{HPF}}(z) = \frac{1}{2} \left(1 - A(z)\right) $.

Here,  $A(z)$ is the transfer function of the first-order **allpass filter** and $c$ the coefficient described above.


Following this logic, we get to the **difference equations**:

**LPF**

$
Y_{\text{LPF}}(n) = \frac{1}{2}(x(n) + A(z))
$

<br>

**HPF**

$
Y_{\text{HPF}}(n) = \frac{1}{2}(x(n) - A(z))
$
<br>

Below an **HPF** example using the Allpass as well as $f_c = fs \cdot 0.1$ of the previous section

```python
fs = sr
# HPF Frequency response
freqs, w, A = allpass_freq_response(c, fs)

hpf_response = 0.5 *(1 - A)
phase_radians = np.angle(hpf_response)
phase_degrees = np.degrees(phase_radians)

# HPF

N = len(samples)
filtered = allpass_first_order(samples, c, fs)

# Applying the HPF in the Allpass output
for n, sample in enumerate(samples):
    filtered[n] = 0.5 *(sample - filtered[n])


def to_db(mags):
    return 10 * np.log10(mags + 1e-20)

plt.figure(figsize=(10,10))

plt.subplot(4, 1, 1)
plt.title("Allpass-based HPF: Magnitude Response")
plt.plot(freqs, to_db(np.abs(hpf_response)))
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude(db)")
plt.xscale('log')
plt.ylim(-25, 10)
plt.axhline(y=-3, linestyle='--', color='red', label="-3db cutoff")
plt.legend()
plt.grid(True)

plt.subplot(4, 1, 2)
plt.title("Allpass-based HPF: Phase Response")
plt.plot(freqs, phase_degrees)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Phase Degrees")
plt.xscale('log')
plt.grid(True)

plt.subplot(4, 1, 3)
plt.title("Original Samples")
plt.plot(np.linspace(0, N, N), samples)
plt.xlabel("Samples")
plt.ylabel("Amplitude")

plt.subplot(4, 1, 4)
plt.title("filtered Samples")
plt.plot(np.linspace(0, N, N), filtered)
plt.xlabel("Samples")
plt.ylabel("Amplitude")

plt.tight_layout()

print("Original")
display(Audio(samples,rate=sr))
print("Filtered")
display(Audio(filtered, rate=sr))
```

### **Second-Order Allpass**

A second-order allpass filter serves as the foundation for creating **bandpass** and **bandreject** filters. To understand how, we first need to examine the second-order allpass filter itself. The **difference equation** is given by:

$$
\large
x_h(n) = x(n) - d(1 - c) \cdot x_h(n - 1) + c \cdot x_h(n - 2)
\large
$$

$$
\large
y(n) = -c \cdot x_h(n) + d(1 - c) \cdot x_h(n - 1) + x_h(n - 2)
\large
$$

With the corresponding coefficients $c$:

$$
\large
c = \frac{tan(\pi \cdot f_b / f_s) - 1 }{tan(\pi \cdot f_b / f_s) + 1}
\large
$$


And $d$:

$$
\large
d = -cos(2 \pi \cdot f_c/f_s)
\large
$$
<br>

Where $f_b$ is the **bandwidth**, controlling the phase transition slope, and $f_c$ is the **cutoff frequency**.

The **transfer function** is:

$$
\large
 A(z) = \frac{-c + d(1 - c)\cdot z^{-1} + z^{-2}}{1 + d(1 - c)\cdot z^{-1} - c \cdot z^{-2}}
 \large
$$

The coefficient $d$ adjusts the center frequency, and $c$ controls the bandwidth. The magnitude response remains constant (equal to 1), while the phase response approaches $-360^\circ$ at high frequencies.

Below an example implementation.



```python
def coeff_c(fb, fs):
    tan = np.tan(np.pi * fb / fs)
    
    return (tan - 1) / (tan + 1)

def coeff_d(fc, fs):
    return -np.cos(2 * np.pi * fc / fs)

def second_order_allpass(x, c, d):
    N = len(x)
    x_h = np.zeros(N)
    y = np.zeros(N)

    for n in range(N):
        if n == 0:
            x_h[n] = x[n] # - d(1 - c) * 0 + c * 0
            y[n] = -c * x_h[n] # + d(1 - c) * 0 +  0
            continue

        if n == 1:
            x_h[n] = x[n] - d * (1 - c) * x_h[n - 1] # + c * 0
            y[n] = -c * x_h[n] + d * (1 - c) * x_h[n - 1] # + 0
            continue
            
        x_h[n] = x[n] - d * (1 - c) * x_h[n - 1] + c * x_h[n - 2]
        y[n] = -c * x_h[n] + d * (1 - c) * x_h[n - 1] + x_h[n - 2]

    return y

def second_order_allpass_freq_response(c, d, n_points=2048):
    w = np.linspace(0, np.pi, n_points)
    A = np.zeros_like(w, dtype=complex)

    for n, omega in enumerate(w):
        z = np.exp(1j * omega)
        num = -c + d * ( 1 - c) * z**-1 + z**-2
        den = 1 + d * ( 1- c) * z**-1 -  c * z**-2
        
        A[n] = num / den

    freqs = (w * fs)/ (2 * np.pi)
    return freqs, w, A

n_samples = 100
fc = 0.1 * fs # hz
fb = 0.022 * fs # bandwdith
fs = sr # sampling frequency (this is correct, don[t worry])
c = coeff_c(fb, fs)
d = coeff_d(fc, fs)

segment = samples[n_samples:n_samples + n_samples] # first 100 samples, this is an allpass, we want to check the phase difference in the time domain
filtered = second_order_allpass(segment, c, d)

freqs, w, A = second_order_allpass_freq_response(c, d)

# unwrapping is necessary because the phase shift in second-order approaches -360 degrees, so we must unwrap as
# our angle is between -pi and pi
phase_radians = np.unwrap(np.angle(A))
phase_degrees = np.degrees(phase_radians)
group_delay = -np.diff(phase_radians) / np.diff(w)

plt.figure(figsize=(10, 10))
plt.subplot(5, 1, 1)
plt.title("Second-order Allpass: Magnitude Response")
plt.plot(freqs, np.abs(A))
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.xlim(0, 20000)
plt.grid(True)

plt.subplot(5, 1, 2)
plt.title("Second-roder Allpass: Phase Response")
plt.plot(freqs, phase_degrees)
plt.xlabel("Frequency(Hz)")
plt.ylabel("Phase (Degrees)")
plt.xlim(0, 20000)
plt.grid(True)

plt.subplot(5, 1, 3)
plt.title("Second-order Allpass: Group Delay")
plt.plot(freqs[:-1], group_delay)
plt.xlabel("Frequency(Hz)")
plt.ylabel("Group Delay (Samples)")
plt.xlim(0, 20000)
plt.grid(True)

plt.subplot(5, 1, 4)
plt.title("Original: 100 samples")
plt.plot(np.linspace(0, n_samples, n_samples), segment)
plt.xlabel("Samples")
plt.ylabel("Amplitude")

plt.subplot(5, 1, 5)
plt.title("Filtered: 100 samples")
plt.plot(np.linspace(0, n_samples, n_samples), filtered)
plt.xlabel("Samples")
plt.ylabel("Amplitude")

plt.tight_layout()
```

### **Second-Order Allpass-based Filter: Bandpass & Bandreject**

A second-order allpass filter can be used to implement **bandpass** and **bandreject** filters by subtracting or adding the output of the allpass filter from/to the input signal.

- **Bandpass filter**: Subtract the allpass filter output from the input signal. This causes **phase cancellation** at frequencies outside the passband, allowing the frequencies within the passband to pass through.
  
- **Bandreject filter**: Add the allpass filter output to the input signal. This causes **phase cancellation** in the passband, allowing frequencies outside the passband to pass through, effectively rejecting the band of interest.

The **transfer function** for these filters is given by:

$$
\large
H(z) = \frac{1}{2} \left(1 \pm A(z)\right)
\large
$$

Where:
- The **subtraction** $(1 - A(z))$ is for the **bandpass filter**.
- The **addition**  $(1 + A(z))$ is for the **bandreject filter**.

Here, $ A(z)$ is the transfer function of the **second-order allpass filter**, which introduces a phase shift that facilitates the filtering effect.

**Bandpass Filter Difference Equation**:

$
\large
Y_{\text{BPF}}(n) = \frac{1}{2} \left(x(n) - A(z)\right)
\large
$

**Bandreject Filter Difference Equation**:

$
\large
Y_{\text{BRF}}(n) = \frac{1}{2} \left(x(n) + A(z)\right)
\large
$

Below an example implementation. Note that the second-order **Allpass from the previous section** is being used here.

```python
fc = 3000 # cutoff in hz 
fb = 0.022 * fs # bandwdith
fs = sr 
c = coeff_c(fb, fs)
d = coeff_d(fc, fs)

filtered = second_order_allpass(samples, c, d)
filtered = 0.5 * (samples - filtered)

freqs, w, A = second_order_allpass_freq_response(c,d)

H_z = 0.5 * (1 - A)
phase_radians = np.unwrap(np.angle(H_z))
phase_degrees= np.degrees(phase_radians)
group_delay = -np.diff(phase_radians) / np.diff(w)

plt.figure(figsize=(6,4))
plt.plot(np.linspace(0, len(filtered), len(filtered)), filtered)
plt.title("Filtered Audio: Bandpass @ 3k")
plt.xlabel("Samples")
plt.ylabel("Amplitude")

plt.figure(figsize=(10, 10))
plt.subplot(5, 1, 1)
plt.title("BPF: Magnitude Response")
plt.plot(freqs, np.abs(H_z))
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.xlim(0, 20000)
plt.grid(True)

plt.subplot(5, 1, 2)
plt.title("BPF: Phase Response")
plt.plot(freqs, phase_degrees)
plt.xlabel("Frequency(Hz)")
plt.ylabel("Phase (Degrees)")
plt.xlim(0, 20000)
plt.grid(True)

plt.subplot(5, 1, 3)
plt.title("BPF: Group Delay")
plt.plot(freqs[:-1], group_delay)
plt.xlabel("Frequency(Hz)")
plt.ylabel("Group Delay (Samples)")
plt.xlim(0, 20000)
plt.ylim(0, 15)
plt.grid(True)

plt.tight_layout()


print("Filtered: Bandpass @ 3k")
display(Audio(filtered, rate=fs))
```
