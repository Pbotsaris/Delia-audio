# **Parametric High-Order Shelving Filter**

This implementation is based on the work of Holters and Zölzer [link here](../papers/holters+zolzer-parametric_high_order_shelving_filters.pdf), who developed an efficient approach for realizing high-order shelving filters with flexible parameter control.

The foundation of this implementation is a low-shelving filter that can be transformed into other filter types (high-shelf, band-shelf) through transformations. The key innovation is the filter's structure: it uses parallel processing paths that allow for efficient computation while maintaining the flexibility to increase to any arbitrary filter order $M$.

## **Filter Structure**
The basic building block is a second-order section described by:

$\Large \overline{H}^{(m)}_{LS,M}(z) = 1 + 2V\frac{1 + c_m}{z^2 + 2c_mz + 1} + V^2\frac{1}{z^2 + 2c_mz + 1}$

where:
- $V$ controls the gain. 
- $c_m$ determines the filter's frequency response

More on the pameters below.

### **Cascading Sections**
The complete filter combines multiple sections through cascading:

$\Large H_{LS,M}(z) = H^{(\frac{M+1}{2})}_{LS,M}(z) \cdot \prod_{m=1}^{\lfloor M/2 \rfloor} \overline{H}^{(m)}_{LS,M}(z)$

This equation shows how we build higher-order filters by multiplying (cascading) individual sections together:

For even orders (M = 2, 4, 6...):
  - We only use second-order sections
  - Example: M = 4 uses two second-order sections

For odd orders (M = 3, 5, 7...):
  - We use second-order sections plus one first-order section
  - Example: M = 3 uses one second-order section plus one first-order section

The term $\lfloor M/2 \rfloor$ (floor division) tells us how many second-order sections we need:
- For M = 3: $\lfloor 3/2 \rfloor = 1$ second-order section
- For M = 4: $\lfloor 4/2 \rfloor = 2$ second-order sections

#### **Parameter $V$ (Gain Control)**

**$V$** is the **gain parameter**, which scales the filter’s frequency response. It is calculated as:

$
\large
V = \sqrt[M]{g} - 1
\large
$

where $g$ is the **linear gain** and $M$ is the filter order. If gain $G$ is given in decibels (dB), convert it to linear gain using:

$
\large
g = 10^{G / 20}
\large
$

then, for a **second-order** filter, $V$ becomes:

$
\large
V = \sqrt{10^{G / 20}} - 1
\large
$

#### **Parameter  $K$ (Bandwidth Control)**

**$K$** is a **bandwidth-related parameter** that defines the frequency range around the cutoff frequency. It is given by:

$
\large
K = \tan\left(\frac{\Omega_b}{2}\right)
\large
$

Where  $\Omega_B $ is the bandwidth in radians per sample and is derived from the cut-off frequency $f_c$ and sampling rate $f_s$:

$
\large
\Omega_b = \frac{2\pi f_c}{f_s}
$

Because this design is based on converting from a continuous time filter, The warping factor $K = tan(\pi f_c/f_s)$ is crucial to ensure the cutoff frequency of the digital filter matches the desired analog cutoff frequency.

$K $ is independent of the filter order  $M $, but as  $M $ increases, additional terms in the transfer function sharpen the filter’s transition around the cutoff.

<a id="cosine-terms"></a>
#### **Parameter  $c_m $ (Cosine Terms for Pole-Zero Placement)**

**$c_m$** is a cosine term that determines the placement of poles and zeros for each section, enhancing the shelving effect with higher filter orders. For an arbitrary order  $M $,  $c_m $ varies by section  $m $ and is calculated as:

$$
\large
c_m = \cos\left(\alpha_m\right) \quad \text{where} \quad \alpha_m = \left(\frac{1}{2} - \frac{2m - 1}{2M}\right) \pi
\large
$$

- The angle  $\alpha_m $ adjusts based on the section index  $m $ and the order  $M $.
- This calculation provides different values for  $c_m $ across filter sections, producing a steeper slope and more pronounced shelving effect as the order  $M $ increases.



## **Second-order Low Shelf**
We'll start by examining a single second-order low shelf filter, which serves as the fundamental building block for higher-order implementations. A second-order section ($M = 2$) represents the simplest complete implementation of the parallel path structure described earlier.

### **From Generic to Specific Form**
The second-order low shelf filter can be derived from the generic continuous-time form:

$\large H_{LS,2}(s) = \frac{s^2 + \sqrt{2}gs + g}{s^2 + \sqrt{2}s + 1}$

After applying the bilinear transform with frequency warping $s = K(1-z^{-1})/(1+z^{-1})$, and factoring the result into parallel paths, we get:
<br><br>

$$ \large H_{LS,2}(z) = 1 + 2VK \frac{K + c + 2Kz^{-1} + (K - c)z^{-2}}{1 + 2Kc + K^2 + (2K^2 - 2)z^{-1} + (1 - 2Kc + K^2)z^{-2}} + V^2 K^2 \frac{1 + 2z^{-1} + z^{-2}}{1 + 2Kc + K^2 + (2K^2 - 2)z^{-1} + (1 - 2Kc + K^2)z^{-2}}$$
<br><br>


This transfer function decomposes into three parallel paths:
1. Direct path: Unity gain (the "1" term)
2. First scaled path: Scaled by V (the middle term)
3. Second scaled path: Scaled by V² (the last term)

where:
- $K = tan(\pi f_c/f_s)$ is the frequency warping factor
- $V = \sqrt{10^{G/20}} - 1$ controls the gain
- $c = cos(\pi/4)$ for the second-order case

### **Building Higher Orders**
Higher-order filters can be constructed by cascading multiple second-order sections. For example, a fourth-order filter would be:

$\large H_{LS,4}(z) = H_{LS,2}(z) \cdot H_{LS,2}(z)$

Each section would have its own cm value calculated according to:

$\large \alpha_m = (\frac{1}{2} - \frac{2m-1}{2M})\pi$

This cascaded structure maintains the advantages of the parallel path implementation while allowing for steeper filter responses through higher orders. We will look into high order shelfs later in this notebook.

### **Coefficient Derivation**

#### **Normalization Factor**
First, we compute an intermediate normalization coefficient:

$
\large
a_0^{-1} = \frac{1}{1 + 2Kc + K^2}
$

#### **Feedback Coefficients (shared between paths)**
$
\large
a_1 = 2(K^2 - 1) \cdot a_0^{-1}
$

$
\large
a_2 = (1 - 2Kc + K^2) \cdot a_0^{-1}
$

#### **First Path Feedforward Coefficients**
Scaled by $V \cdot k_1$ where $k_1 = 2K$:

$
\large
b_{0,1} = (K + c) \cdot a_0^{-1}
$

$
\large
b_{1,1} = 2K \cdot a_0^{-1}
$

$
\large
b_{2,1} = (K - c) \cdot a_0^{-1}
$

#### **Second Path Feedforward Coefficients**
Scaled by $V^2 \cdot k_2$ where $k_2 = K^2$:

$
\large
b_{0,2} = a_0^{-1}
$

$
\large
b_{1,2} = 2 \cdot a_0^{-1}
$

$
\large
b_{2,2} = a_0^{-1}
$

Where:
- $K = \tan(\pi f_c/f_s)$ → bilinear transform warping
- $V = \sqrt{10^{G/20}} - 1$ → gain parameter
- $c = \cos(\pi/4)$  → for second-order filter

### **Difference Equations**

The filter processes each path separately and then combines them:

First Path: 

$$
\large
v_1[n] = b_{0,1} \cdot x[n] + b_{1,1} \cdot x[n-1] + b_{2,1} \cdot x[n-2] - a_1 \cdot v_1[n-1] - a_2 \cdot v_1[n-2]
$$

Second Path:

$$
\large
v_2[n] = b_{0,2} \cdot x[n] + b_{1,2} \cdot x[n-1] + b_{2,2} \cdot x[n-2] - a_1 \cdot v_2[n-1] - a_2 \cdot v_2[n-2]
$$

Combine outputs: 

$$
\large
y[n] = x[n] + V \cdot k_1 \cdot v_1[n] + V^2 \cdot k_2 \cdot v_2[n]
$$

### **Second-Order Example**

```python
fc = 500
G = -10

def low_shelf_coeff(G, fc, fs):
    K = np.tan(np.pi * fc / fs)
    V = np.sqrt(10 ** (G / 20)) - 1
    c = np.cos(np.pi / 4)  # For second-order filter
    
    # Calculate the intermediate coefficient \alpha_0^{-1}
    a0_inv = 1 / (1 + 2 * K * c + K**2)
    
    # Coefficients for the recursive part (denominator)
    a1 = 2 * (K**2 - 1) * a0_inv
    a2 = (1 - 2 * K * c + K**2) * a0_inv
    
    # First parallel path coefficients (scaled by V)
    k1 = 2 * K
    b0_1 = (K + c) * a0_inv
    b1_1 = k1 * a0_inv
    b2_1 = (K - c) * a0_inv
    
    # Second parallel path coefficients (scaled by V^2)
    k2 = K**2
    b0_2 = a0_inv
    b1_2 = 2 * a0_inv
    b2_2 = a0_inv
    
    return k1, k2, b0_1, b1_1, b2_1, b0_2, b1_2, b2_2, a1, a2

def low_shelf(x, G, fc, fs):
    N = len(x)
    y = np.zeros(N)
    v1 = np.zeros(N)  # First parallel path output
    v2 = np.zeros(N)  # Second parallel path output
    
    k1, k2, b0_1, b1_1, b2_1, b0_2, b1_2, b2_2, a1, a2 = low_shelf_coeff(G, fc, fs)
    V = np.sqrt(10 ** (G / 20)) - 1
    
    # Initialize first two samples
    v1[0] = b0_1 * x[0]
    v2[0] = b0_2 * x[0]
    y[0] = x[0] + V * k1 * v1[0] + V * V * k2 * v2[0]
    
    v1[1] = b0_1 * x[1] + b1_1 * x[0] - a1 * v1[0] - a2 * v1[0]
    v2[1] = b0_2 * x[1] + b1_2 * x[0] - a1 * v2[0] - a2 * v2[0]
    y[1] = x[1] + V * k1 * v1[1] + V * V * k2 * v2[1]
    
    # Process remaining samples
    for n in range(2, N):
        # First parallel path
        v1[n] = (b0_1 * x[n] + b1_1 * x[n-1] + b2_1 * x[n-2] - 
                 a1 * v1[n-1] - a2 * v1[n-2])
        
        # Second parallel path
        v2[n] = (b0_2 * x[n] + b1_2 * x[n-1] + b2_2 * x[n-2] - 
                 a1 * v2[n-1] - a2 * v2[n-2])
        
        # Combine paths with scaling
        y[n] = x[n] + V * k1 * v1[n] + V * V * k2 * v2[n]
    
    return y


def low_shelf_freq_response(G, fc, fs, n_points=1024):
    w = np.linspace(0, np.pi, n_points)
    H = np.zeros_like(w, dtype=complex)
    K = np.tan(np.pi * fc / fs)
    V = np.sqrt(10 ** (G / 20)) - 1
    c = np.cos(np.pi / 4)

    for i, omega in enumerate(w):
        z = np.exp(1j * omega)

        # First term
        num1 = 2 * V * K * (K + c + 2 * K * z**-1 + (K - c) * z**-2)
        den = 1 + 2 * K * c + K**2 + (2 * K**2 - 2) * z**-1 + (1 - 2 * K * c + K**2) * z**-2

        # Second term
        num2 = V**2 * K**2 * (1 + 2 * z**-1 + z**-2)

        # Combined transfer function with direct path term
        H[i] = 1 + (num1 / den) + (num2 / den)


    freqs = (w * fs) / (2 * np.pi)
    return freqs, 20 * np.log10(np.abs(H))


filtered = low_shelf(samples, G, fc, fs)

freqs, mags = low_shelf_freq_response(G,fc, fs)

plt.figure(figsize=(16, 6))
plt.plot(freqs, mags, label=f"Low Shelf Filter (G={G} dB, fc={fc} Hz)")
plt.xscale('log')
plt.xlim([20, fs / 2])
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude (dB)")
plt.title("Frequency Response of Low Shelf Filter")
plt.grid(True, which="both", ls="--", lw=0.5)
plt.legend()
plt.show()

print("Original")
display(Audio(samples, rate=fs))
print(f"Second-order Shelf fs={fs}, G={G} ")
display(Audio(filtered, rate=fs)) 
```

## **Even Order Low Shelf**

Having implemented the second-order low shelf filter, we can now extend our implementation to handle arbitrary filter orders (M). This is achieved by cascading multiple second-order sections, where each section uses a different set of coefficients calculated from its position in the cascade.

For a filter of order M, we need:
- $M/2$ second-order sections for even orders (e.g., M = 4 uses two sections)
- $M/2$ second-order sections plus one first-order section for odd orders (e.g., M = 3 uses one section of each)

Each second-order section maintains the parallel path structure we saw earlier, but uses a unique cm value calculated as:

$\large \alpha_m = (\frac{1}{2} - \frac{2m-1}{2M})\pi$

where m is the section number (1 to M/2).

Let's start with the implementation for even orders, which is simpler as it only requires second-order sections. We'll modify our previous implementation in three key ways:
1. Calculate coefficients for multiple sections
2. Process the input through each section sequentially
3. Combine the responses of all sections for the final output

```python
G = -10
fc = 600
M = 6

def low_shelf_coeffs(G, fc, fs, M):
    K = np.tan(np.pi * fc / fs)
    V = np.sqrt(10 **(G/20)) - 1

    coeffs = []

    for m in range(1, M // 2 + 1):
        # we must calculate the cm for each section
        alpha_m = (1 / 2 - (2 * m - 1) / (2 * M)) * np.pi
        c = np.cos(alpha_m)

        a0_inv = 1 / (1 + 2 * K * c + K**2)

        # coeffs for recursive part (denominator)
        a1 = 2 * (K**2 - 1) * a0_inv
        a2 = (1 - 2 * K * c + K**2) * a0_inv

        # first parallel path (scaled by V)
        k1 = 2 * K
        b0_1 = (K +c) * a0_inv
        b1_1 = k1 * a0_inv
        b2_1 = (K - c) * a0_inv

        # second parellel path (scaled by V^2)
        k2 = K**2 

        b0_2 = a0_inv
        b1_2 = 2 * a0_inv
        b2_2 = a0_inv

        coeffs.append((k1, k2, b0_1, b1_1, b2_1, b0_2, b1_2, b2_2, a1, a2))

    return coeffs


def generic_low_shelf(x, G, fc, fs, M):
    N = len(x)
    # this is important because we will compute the filter multiple times
    y = x.copy()

    coeffs = low_shelf_coeffs(G,fc, fs, M)
    V = np.sqrt( 10 ** (G / 20)) - 1

    for coeff in coeffs:
        k1, k2, b0_1, b1_1, b2_1, b0_2, b1_2, b2_2, a1, a2 = coeff

        v1 = np.zeros(N) # first parallel path
        v2 = np.zeros(N) # second parallel path
        section_out = np.zeros(N)

        # first sample
        v1[0] = b0_1 * y[0]
        v2[0] = b0_2 * y[0]
        section_out[0] = y[0] + V * k1 * v1[0] + V * V * k2 * v2[0]

        # second sample
        v1[1] = b0_1 * y[1] + b1_1 * y[0] - a1 * v1[0]  - a2 * v1[0]
        v2[1] = b0_2  * y[1] + b1_2 * y[0] - a1 * v2[0] - a2 * v2[0]
        section_out[1] = y[1] + V  * k1 * v1[1] + V * V * k2 * v2[1] 

        for n in range(2, N):
            v1[n] = (b0_1 * y[n] + b1_1 * y[n - 1] + b2_1 * y[n - 2] - a1 * v1[n - 1] - a2 * v1[n - 2])
            v2[n] = (b0_2 * y[n] + b1_2 * y[n - 1] + b2_2 * y[n - 2] - a1 * v2[n - 1] - a2 * v2[n - 2])

            section_out[n] = y[n] + V * k1 * v1[n] + V * V * k2 * v2[n]

        y = section_out.copy()


    return y

def generic_low_shelf_freq_response(G, fc, fs, M, n_points=1024):
    w = np.linspace(0, np.pi, n_points)
    H = np.ones_like(w, dtype=complex)
    K = np.tan(np.pi * fc / fs)
    V = np.sqrt(10 ** (G / 20)) - 1

    # acummulate for each section
    for m in range(1, M // 2 + 1):
        alpha_m = (1 / 2 - (2 * m - 1) / (2 * M)) * np.pi
        c = np.cos(alpha_m)

        for i, omega in enumerate(w):
            z = np.exp(1j * omega)
    
            num1 = 2 * V * K * (K + c + 2 * K * z**-1 + (K - c) * z**-2)
            den = 1 + 2 * K * c + K**2 + (2 * K**2 - 2) * z**-1 + (1 - 2 * K * c + K**2) * z**-2
            
            num2 = V**2 * K**2 * (1 + 2 * z**-1 + z**-2)
    
            H[i] *= 1 + (num1 / den) +  (num2 / den)


    # we must normalize as we have multiple sections
    H = H ** (1 / (M // 2))
    
    freqs = ( w * fs) / (2 * np.pi)

    return freqs, 20 * np.log10(np.abs(H))
    

filtered = generic_low_shelf(samples, G, fc, fs, M)


freqs, mags = generic_low_shelf_freq_response(G, fc, fs, M)

plt.figure(figsize=(16, 6))
plt.plot(freqs, mags, label=f"Low Shelf Filter (M={M}, G={G} dB, fc={fc} Hz)")
plt.xscale('log')
plt.xlim([20, fs / 2])
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude (dB)")
plt.title("Freq Response Low Shelf Filter")
plt.grid(True, which="both", ls="--", lw=0.5)
plt.legend()
plt.show()


print("Original")
display(Audio(samples, rate=fs))

print(f"{M}th-order filter fc={fc}, G={G}")
display(Audio(filtered , rate=fs))    
```

## **Arbitrary Order Low Shelf**

Expanding our implementation to handle any order M requires special consideration for odd-ordered filters. When M is odd, we need to include an additional first-order section alongside the second-order sections.

The first-order low-shelf transfer function is given by:

$\large H_{LS,1}(s) = \Large \frac{s + g}{s + 1}$

After applying the bilinear transform, this becomes:

$\large H_{LS,1}(z) = 1 + \frac{V}{2}(1 + \Large \frac{a + z^{-1}}{1 + az^{-1}})$

where:
- $a = \frac{K-1}{K+1}$ is the all-pass coefficient
- $K = \tan(\pi f_c/f_s)$ is the frequency warping factor
- $V = \sqrt{10^{G/20}} - 1$ controls the gain

The first-order difference equation is:

$\large y[n] = x[n] + \frac{V}{2}(x[n] + v[n])$

where $v[n]$ is the all-pass filter output:

$\large v[n] = a \cdot v[n-1] + x[n] - a \cdot x[n-1]$

Our implementation handles both even and odd orders through three main functions:
1. `low_shelf_coeffs`: Calculates coefficients for both first- and second-order sections based on M
2. `generic_low_shelf`: Processes the signal through cascaded sections appropriately
3. `generic_low_shelf_freq_response`: Computes the complete frequency response including first-order sections when M is odd

For odd orders (e.g., M = 3), the signal flows through ⌊M/2⌋ second-order sections followed by one first-order section. For even orders (e.g., M = 4), only second-order sections are used.

```python
G = 10
fc = 600
M = 9

def low_shelf_coeffs(G, fc, fs, M):
    K = np.tan(np.pi * fc / fs)
    V = np.sqrt(10 **(G/20)) - 1

    coeffs = []

    for m in range(1, M // 2 + 1):
        alpha_m = (1 / 2 - (2 * m - 1) / (2 * M)) * np.pi
        c = np.cos(alpha_m)

        a0_inv = 1 / (1 + 2 * K * c + K**2)

        # coeffs for recursive part (denominator)
        a1 = 2 * (K**2 - 1) * a0_inv
        a2 = (1 - 2 * K * c + K**2) * a0_inv

        # first parallel path (scaled by V)
        k1 = 2 * K
        b0_1 = (K +c) * a0_inv
        b1_1 = k1 * a0_inv
        b2_1 = (K - c) * a0_inv

        # second parellel path (scaled by V^2)
        k2 = K**2 

        b0_2 = a0_inv
        b1_2 = 2 * a0_inv
        b2_2 = a0_inv

        # coeffs the either `first` or `second` order 
        coeffs.append( ('second', (k1, k2, b0_1, b1_1, b2_1, b0_2, b1_2, b2_2, a1, a2)))

    # Calculate first-order coeffs
    if M % 2 == 1:
        a = (K - 1) / (K + 1)
        coeffs.append((('first'), (a, K)))

    return coeffs


def generic_low_shelf(x, G, fc, fs, M):
    N = len(x)
    y = x.copy()

    coeffs = low_shelf_coeffs(G,fc, fs, M)
    V = np.sqrt( 10 ** (G / 20)) - 1

    for section_type, coeff in coeffs:
        # First-order
        if section_type == 'first':
            a, K = coeff

            v = np.zeros(N)
            section_out = np.zeros(N)
            v[0] = y[0]
            section_out[0] = y[0] + (V / 2) * (y[0] + v[0])
            
            for n in range(1, N):
                v[n] = a * v[n - 1] + y[n] - a * y[n - 1]  
                section_out[n] = y[n] + (V / 2) * (y[n] + v[n])  
            y = section_out.copy()
            continue

        # Second-order
        k1, k2, b0_1, b1_1, b2_1, b0_2, b1_2, b2_2, a1, a2 = coeff

        v1 = np.zeros(N) 
        v2 = np.zeros(N) 
        section_out = np.zeros(N)

        v1[0] = b0_1 * y[0]
        v2[0] = b0_2 * y[0]
        section_out[0] = y[0] + V * k1 * v1[0] + V * V * k2 * v2[0]

        v1[1] = b0_1 * y[1] + b1_1 * y[0] - a1 * v1[0] - a2 * v1[0] 
        v2[1] = b0_2  * y[1] + b1_2 * y[0] - a1 * v2[0] - a2 * v2[0]
        section_out[1] = y[1] + V  * k1 * v1[1] + V * V * k2 * v2[1] 

        for n in range(2, N):
            v1[n] = (b0_1 * y[n] + b1_1 * y[n - 1] + b2_1 * y[n - 2] - a1 * v1[n - 1] - a2 * v1[n - 2])
            v2[n] = (b0_2 * y[n] + b1_2 * y[n - 1] + b2_2 * y[n - 2] - a1 * v2[n - 1] - a2 * v2[n - 2])
            section_out[n] = y[n] + V * k1 * v1[n] + V * V * k2 * v2[n]

        y = section_out.copy()

    return y

def generic_low_shelf_freq_response(G, fc, fs, M, n_points=1024):
    w = np.linspace(0, np.pi, n_points)
    H = np.ones_like(w, dtype=complex)
    K = np.tan(np.pi * fc / fs)
    V = np.sqrt(10 ** (G / 20)) - 1

    # acummulate for each section
    for m in range(1, M // 2 + 1):
        alpha_m = (1 / 2 - (2 * m - 1) / (2 * M)) * np.pi
        c = np.cos(alpha_m)

        for i, omega in enumerate(w):
            z = np.exp(1j * omega)
    
            num1 = 2 * V * K * (K + c + 2 * K * z**-1 + (K - c) * z**-2)
            den = 1 + 2 * K * c + K**2 + (2 * K**2 - 2) * z**-1 + (1 - 2 * K * c + K**2) * z**-2
            
            num2 = V**2 * K**2 * (1 + 2 * z**-1 + z**-2)
    
            H[i] *= 1 + (num1 / den) +  (num2 / den)

    # handle the first-order if applicable
    if M % 2 == 1:
        for i, omega in enumerate(w):
            z = np.exp(1j * omega)
            a = (K - 1) / (K + 1)
            # Allpass transfer function
            H_ap = (a + z**-1) / (1 + a * z**-1)
            H[i] *= 1 + (V / 2) * (1 + H_ap)
            
    H = H ** (1 / (M // 2))
    freqs = ( w * fs) / (2 * np.pi)

    return freqs, 20 * np.log10(np.abs(H))
    
filtered = generic_low_shelf(samples, G, fc, fs, M)

Ms = [2, 4, 8]

plt.figure(figsize=(16, 8))

for M in Ms:
    freqs, mags = generic_low_shelf_freq_response(G, fc, fs, M)
    plt.semilogx(freqs, mags, label=f'M={M}')

plt.grid(True, which="both", ls="--", lw=0.5)
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axhline(y=G, color='r', linestyle='--', alpha=0.5, label='Target gain')
plt.axvline(x=fc, color='g', linestyle='--', alpha=0.5, label='Cutoff freq')
plt.xlim([20, fs/2])
plt.ylim([min(G-5, -5), max(5, G+5)])
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude (dB)")
plt.title("Low Shelf Filter Frequency Response")
plt.legend()
plt.show()


print("Original")
display(Audio(samples, rate=fs))

print(f"{M}th-order filter fc={fc}, G={G}")
display(Audio(filtered , rate=fs))    
```

## **High-Shelf Transformation**

The high-shelf filter is derived from the low-shelf filter by substituting  $z^{-1} $ with  $-z^{-1} $ in the transfer function. This substitution mirrors the frequency response around  $f_s / 4 $, creating a gain adjustment that affects frequencies above a specified cutoff frequency. The high-shelf transformation is expressed as:

$
\large H_{HF, M}(z) = H_{LF, M}(-z)
$

In practice, this transformation involves:
1. For second-order sections: Changing the sign of terms with odd powers of  $z $.
2. For the first-order section: Replacing  $z^{-1} $ with  $-z^{-1} $.

**Note**: We flip the gain  $G $ by setting  $V = \sqrt{10^{(-G / 20)}} - 1 $ to ensure that positive values of  $G $ boost high frequencies, while negative values apply attenuation.

### **Coefficients**

For each filter section, we derive coefficients based on the high-shelf transformation:

1. **Analog Prototype Transformation**: We set

   $
   \large
   K_{\text{high}} = \frac{1}{K}
   $

   where  $K = \tan\left(\frac{\pi f_c}{f_s}\right) $

2. **Inverse Normalization Factor  $a0_{\text{inv}} $**:

     $
     \large
     a_{0,\text{inv}} = \frac{1}{1 + 2 K_{\text{high}} c + K_{\text{high}}^2}
     $

3. **Feedback Coefficients**: recursive part (denominator)

     $
      \large
     a_1 = -2 (K_{\text{high}}^2 - 1) \cdot a0_{\text{inv}}
     $
     
     $
     \large
     a_2 = (1 - 2 K_{\text{high}} c + K_{\text{high}}^2) \cdot a0_{\text{inv}}
     $

4. **Feedforward Coefficients** for the First Path:

     $
     \large
     b{0,1} = (K_{\text{high}} + c) \cdot a0_{\text{inv}}
     $
     
     $
     \large
     b{1,1} = -2 K_{\text{high}} \cdot a0_{\text{inv}}
     $
     
     $
     \large
     b{2,1} = (K_{\text{high}} - c) \cdot a0_{\text{inv}}
     $

5. **Feedforward Coefficients**: for the Second Path

     $
     \large
     k_2 = K_{\text{high}}^2
     $
     
     $
     \large
     b{0,2} = a0_{\text{inv}}
     $
     
     $
     \large
     b{1,2} = -2 \cdot a0_{\text{inv}}
     $
     
     $
     \large
     b{2,2} = a0_{\text{inv}}
    $


### **Note on High-Shelf Filter Frequency Response**
In the implementation below we are normalizing the frequency response with an arbitrary factor of `1.58`. This is not theoretically ideal, but it resolves discrepancies that arise when transforming the high-shelf filter from analog to digital domain. 

The primary challenge lies in the correct implementation of the frequency transformation from low-shelf to high-shelf, where $z^{-1}$ becomes $-z^{-1}$. While the paper by Holters and Zölzer primarily discusses these transformations in continuous time, deriving the correct discrete-time transfer function proved challenging for the high-shelf case.

For comparison, the band-shelf transformation was successfully derived for discrete-time (see implementation below), but the high-shelf transformation remains an open problem for future investigation. This suggests there might be subtleties in the bilinear transform and frequency inversion that we're not fully accounting for in our frequency response calculation.

The current normalization factor of `1.58` in `H = H ** (1.58 / (M // 2))` approximates the expected behavior, but a more theoretically sound solution would be preferable.

```python
G = 10
fc = 1000
M = 8

def high_shelf_coeffs(G, fc, fs, M):
    K = np.tan(np.pi * fc / fs)
    # for whatever reason I need to invert the G here
    V = np.sqrt(10 ** (-G / 20)) - 1
    
    coeffs = []
    
    for m in range(1, M // 2 + 1):
        alpha_m = (1 / 2 - (2 * m - 1) / (2 * M)) * np.pi
        c = np.cos(alpha_m)
        
        K_high = 1 / K  # Analog prototype transformation
        a0_inv = 1 / (1 + 2 * K_high * c + K_high**2)
        
        # High-shelf coefficients with -z^(-1)
        a1 = -2 * (K_high**2 - 1) * a0_inv
        a2 = (1 - 2 * K_high * c + K_high**2) * a0_inv
        
        k1 = -2 * K_high
        b0_1 = (K_high + c) * a0_inv
        b1_1 = k1 * a0_inv
        b2_1 = (K_high - c) * a0_inv
        
        k2 = K_high**2
        b0_2 = a0_inv
        b1_2 = -2 * a0_inv
        b2_2 = a0_inv
        
        coeffs.append(('second', (k1, k2, b0_1, b1_1, b2_1, b0_2, b1_2, b2_2, a1, a2)))

    if M % 2 == 1:
        K_high = 1 / K
        a = (K_high - 1) / (K_high + 1)
        coeffs.append(('first', (a, K_high)))
        
    return coeffs, V

def high_shelf_filter(x, G, fc, fs, M):
    N = len(x)
    y = x.copy()
    coeffs, V = high_shelf_coeffs(G, fc, fs, M)

    for section_type, coeff in coeffs:
        # First-order
        if section_type == 'first':
            a, K = coeff

            v = np.zeros(N)
            section_out = np.zeros(N)
            v[0] = y[0]
            section_out[0] = y[0] + (V / 2) * (y[0] + v[0])
            
            for n in range(1, N):
                v[n] = a * v[n - 1] + y[n] - a * y[n - 1]  
                section_out[n] = y[n] + (V / 2) * (y[n] + v[n])  
            y = section_out.copy()
            continue

        # Second-order
        k1, k2, b0_1, b1_1, b2_1, b0_2, b1_2, b2_2, a1, a2 = coeff

        v1 = np.zeros(N) 
        v2 = np.zeros(N) 
        section_out = np.zeros(N)

        v1[0] = b0_1 * y[0]
        v2[0] = b0_2 * y[0]
        section_out[0] = y[0] + V * k1 * v1[0] + V * V * k2 * v2[0]

        v1[1] = b0_1 * y[1] + b1_1 * y[0] - a1 * v1[0] - a2 * v1[0] 
        v2[1] = b0_2  * y[1] + b1_2 * y[0] - a1 * v2[0] - a2 * v2[0]
        section_out[1] = y[1] + V  * k1 * v1[1] + V * V * k2 * v2[1] 

        for n in range(2, N):
            v1[n] = (b0_1 * y[n] + b1_1 * y[n - 1] + b2_1 * y[n - 2] - a1 * v1[n - 1] - a2 * v1[n - 2])
            v2[n] = (b0_2 * y[n] + b1_2 * y[n - 1] + b2_2 * y[n - 2] - a1 * v2[n - 1] - a2 * v2[n - 2])
            section_out[n] = y[n] + V * k1 * v1[n] + V * V * k2 * v2[n]

        y = section_out.copy()

    
    return y

def high_shelf_freq_response(G, fc, fs, M, n_points=1024):
    coeffs, V = high_shelf_coeffs(G, fc, fs, M)
    
    w = np.linspace(0, np.pi, n_points)
    freqs = w * fs / (2 * np.pi)
    H = np.ones(n_points, dtype=complex)
    
    for section_type, coeff in coeffs:
        if section_type == 'first':
            a, K = coeff
            z = np.exp(1j * w)
            section_H = 1 + (V/2) * (1 + (1 + z**-1)/(1 + a * z**-1))
            H *= section_H
            
        else:
            k1, k2, b0_1, b1_1, b2_1, b0_2, b1_2, b2_2, a1, a2 = coeff
            z = np.exp(1j * w)
            
            num1 = b0_1 + b1_1 * z**-1 + b2_1 * z**-2
            num2 = b0_2 + b1_2 * z**-1 + b2_2 * z**-2
            den = 1 + a1 * z**-1 + a2 * z**-2
            
            section_H = 1 + V * k1 * (num1/den) + V**2 * k2 * (num2/den)
            H *= section_H

    # normalize with an arbritaty 1.58, I know but I couldn't get this to work in the better
    # way given the the paper is mostly in the time continous domain. Maybe something to look into later
    H = H ** (1.58 / (M // 2))
    return freqs, 20 * np.log10(np.abs(H))

Ms = [2, 4, 8]
Gs =  [-10, 10]
plt.figure(figsize=(16, 8))

print("Original")
display(Audio(samples, rate=fs))

for M in Ms:
    
    # actual_gain= verify_filter_gain(G, fc, fs, M)
    freqs, mags = high_shelf_freq_response(G, fc, fs, M)
    
    filtered = high_shelf_filter(samples, G, fc, fs, M)
    print(f"Filtered: G={G} M={M}")
    display(Audio(filtered, rate=fs))
    
    plt.semilogx(freqs, mags, label=f'M={M}')
    

plt.grid(True, which="both", ls="--", lw=0.5)
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axhline(y=G, color='r', linestyle='--', alpha=0.5, label='Target gain')
plt.axvline(x=fc, color='g', linestyle='--', alpha=0.5, label='Cutoff freq')

plt.xlim([20, fs/2])

# plt.ylim([min(G-5, -5), max(5, G+5)])
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude (dB)")
plt.title("High Shelf Filter Frequency Response")
plt.legend()
plt.show()

```

## **Band-Shelf Transformation**

The band-shelf filter applies gain adjustments within a specific frequency band centered around a chosen frequency  $\Omega_0 $ (in radians per sample). This transformation is derived from a low-shelf filter by substituting  $z^{-1} $ in the transfer function with an all-pass transformation  $A(z) $ that focuses the gain on a frequency band.

### **Center Frequency Shift with All-Pass Transformation**

The key transformation involves replacing  $z^{-1} $ with an all-pass function  $A(z) $ that shifts the filter’s response to the target center frequency  $\Omega_0 $:

$
\large
A(z) = \frac{z^{-1} (c_0 - z^{-1})}{1 - c_0 z^{-1}}
$

where  $c_0 = \cos(\Omega_0) $. This substitution enables the gain to be applied specifically around  $\Omega_0 $, creating a band-shelf effect.

### **Analog Prototype Transformation**

To shift the low-shelf filter’s response to a band-shelf configuration, an analog prototype transformation is applied. This involves setting:

$
\large
K_{\text{band}} = \frac{1}{K}
$

where  $K = \tan\left(\frac{\pi f_c}{f_s}\right) $ (or alternatively,  $K = \tan\left(\frac{\Omega_B}{2}\right) $ for a specified bandwidth  $\Omega_B $). This transformation adjusts the filter response to center around the desired frequency band rather than the entire low-frequency range.

### **Deriving the Filter Coefficients**

For each section in the filter, coefficients are calculated based on the transformed parameters:

1. **Inverse Normalization Factor**:
    Ensures that each section is properly scaled.

     $
     \large
     a_{0,\text{inv}} = \frac{1}{1 + 2 K_{\text{band}} \cdot \LARGE c \large + K_{\text{band}}^2}
     $

2. **Feedback Coefficients** : 
    Define the recursive part (denominator) of each filter section.

     $
     \large
     a_1 = -2 (K_{\text{band}}^2 - 1) \cdot a_{0,{\text{inv}}}
     $
     
     $
     \large
     a_2 = (1 - 2 K_{\text{band}} c + K_{\text{band}}^2) \cdot a_{0, \text{inv}}
     $

3. **Feedforward Coefficients** for the First Path

     $
     \large
     b{0,1} = (K_{\text{band}} + c) \cdot a_{0,\text{inv}}
     $
     
     $
     \large
     b{1,1} = -2 K_{\text{band}} \cdot a_{0,\text{inv}}
     $
     
     $
     \large
     b{2,1} = (K_{\text{band}} - c) \cdot a_{0,\text{inv}}
     $

4. **Feedforward Coefficients:** for the Second Path (scaled by $V^2 $)

     $
     \large
     k_2 = K_{\text{band}}^2
     $
     
     $
     \large
     b_{0,2} = a_{0,\text{inv}}
     $
     
     $
     \large
     b_{1_2} = -2 \cdot a_{0,\text{inv}}
     $
     
     $
     \large
     b_{2_2} = a_{0,\text{inv}}
     $

### **Magnitude Response**

The magnitude response of the band-shelf filter is expressed as:

$
\LARGE |H_{BS, M}(e^{j\omega})|^2 = \frac{(c_0 - \cos \omega)^{2M} + (K \sin \omega)^{2M} g^2}{(c_0 - \cos \omega)^{2M} + (K \sin \omega)^{2M}}
$

where:
- **$g = 10^{-G/40}$**: Similarly to the Highpass, we are inverting the gain here
- **$K = \tan(\frac{\pi f_c}{f_s})$**: Controls the bandwidth in terms of the digital cutoff frequency.
- **$c_0 = \cos(\frac{2 \pi f_c}{f_s})$**: Centers the filter response around the digital cutoff frequency $f_c$.



```python
G = 3
fc = 1000
M = 8


def band_shelf_coeffs(G, fc, fs, M, ):
    
    K = np.tan(np.pi * fc / fs)
    V = np.sqrt(10 ** (-G / 20)) - 1

    Omega0 = 2 * np.pi * fc / fs
    
    # Compute the cosine term for the center frequency shift
    c0 = np.cos(Omega0)
    coeffs = []
    
    
    for m in range(1, M // 2 + 1):
        alpha_m = (1 / 2 - (2 * m - 1) / (2 * M)) * np.pi
        c = np.cos(alpha_m)
        
        K_band = 1 / K  # Analog prototype transformation
        a0_inv = 1 / (1 + 2 * K_band * c + K_band**2)
        
        # Calculate the all-pass substitution terms
        a1 = -2 * (K_band**2 - 1) * a0_inv
        a2 = (1 - 2 * K_band * c + K_band**2) * a0_inv
        
        k1 = -2 * K_band
        # Substitute z^{-1} -> A(z) with center frequency c0
        b0_1 = (K_band + c) * a0_inv
        b1_1 = k1 * a0_inv
        b2_1 = (K_band - c) * a0_inv
        
        k2 = K_band**2
        b0_2 = a0_inv
        b1_2 = -2 * a0_inv
        b2_2 = a0_inv
        
        
        coeffs.append(('second',(k1, k2, b0_1, b1_1, b2_1, b0_2, b1_2, b2_2, a1, a2)))

    if M % 2 == 1:
        K_band = 1 / K
        a = (K_band - 1) / (K_band + 1)
        coeffs.append(('first', (a, K_band)))
        
    return coeffs, V

def band_shelf_filter(x, G, fc, fs, M):
    N = len(x)
    y = x.copy()
    coeffs, V = band_shelf_coeffs(G, fc, fs, M)

    for section_type, coeff in coeffs:
        if section_type == 'first':
            a, K = coeff

            v = np.zeros(N)
            section_out = np.zeros(N)
            v[0] = y[0]
            section_out[0] = y[0] + (V / 2) * (y[0] + v[0])
            
            for n in range(1, N):
                v[n] = a * v[n - 1] + y[n] - a * y[n - 1]  
                section_out[n] = y[n] + (V / 2) * (y[n] + v[n])  
            y = section_out.copy()
            continue

        # Second-order
        k1, k2, b0_1, b1_1, b2_1, b0_2, b1_2, b2_2, a1, a2 = coeff

        v1 = np.zeros(N) 
        v2 = np.zeros(N) 
        section_out = np.zeros(N)

        v1[0] = b0_1 * y[0]
        v2[0] = b0_2 * y[0]
        section_out[0] = y[0] + V * k1 * v1[0] + V * V * k2 * v2[0]

        v1[1] = b0_1 * y[1] + b1_1 * y[0] - a1 * v1[0] - a2 * v1[0] 
        v2[1] = b0_2  * y[1] + b1_2 * y[0] - a1 * v2[0] - a2 * v2[0]
        section_out[1] = y[1] + V  * k1 * v1[1] + V * V * k2 * v2[1] 

        for n in range(2, N):
            v1[n] = (b0_1 * y[n] + b1_1 * y[n - 1] + b2_1 * y[n - 2] - a1 * v1[n - 1] - a2 * v1[n - 2])
            v2[n] = (b0_2 * y[n] + b1_2 * y[n - 1] + b2_2 * y[n - 2] - a1 * v2[n - 1] - a2 * v2[n - 2])
            section_out[n] = y[n] + V * k1 * v1[n] + V * V * k2 * v2[n]

        y = section_out.copy()

    return y


def band_shelf_freq_response(G, fc, fs, M, n_points=1024):
    w = np.linspace(0, np.pi, n_points)
    H = np.ones_like(w, dtype=complex)

    c0 = np.cos(2 * np.pi * fc / fs)
    g = 10 ** (G/40) # 40 because of g**2, though this may be hacky

    K = np.tan(np.pi * fc / fs)

    for m in range(1, M // 2 + 1):
        alpha_m = (1 / 2 - (2 * m - 1) / (2 * M)) * np.pi
        c = np.cos(alpha_m)

        for i, omega in enumerate(w):
            
            cos  =( c0 - np.cos(omega)) ** (2 * M) 
            sin = (K * np.sin(omega)) ** (2 * M) 
            H[i] *= (cos + sin * g**2) / (cos + sin)

    H = H ** (1 / (M // 2))

    freqs = (w * fs ) / (np.pi * 2)

    return freqs, 20 * np.log10(np.abs(H))

print("Original")
display(Audio(samples, rate=fs))

plt.figure(figsize=(16, 8))
for M in Ms:
    freqs, mags =band_shelf_freq_response(G, fc, fs, M)
    plt.semilogx(freqs, mags, label=f'M={M}')
    
    filtered = band_shelf_filter(samples, G, fc, fs, M)
    print(f"Band pass: M={M}, G={G}")
    display(Audio(filtered, rate=fs))

plt.grid(True, which="both", ls="--", lw=0.5)
plt.xlim([20, fs/2])
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude (dB)")
plt.title("High Shelf Filter Frequency Response")
plt.legend()
plt.show()
```
