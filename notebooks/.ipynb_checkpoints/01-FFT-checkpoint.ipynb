# FFT

This notebook documents the process of implementing FFT and IFFT as well as some related utility functions. Below the basic formula for DFT and inverse DFT.

**DFT**

$
\large
X_k = \sum_{n=0}^{N-1} x[n] \cdot e^{-j \frac{2\pi}{N} k n}, \quad k = 0, 1, 2, \dots, N-1
\large
$

**IDFT**

$
\large
x_n = \frac{1}{N} \sum_{k=0}^{N-1} X[n] \cdot e^{j \frac{2\pi}{N} k n}, \quad n = 0, 1, 2, \dots, N-1
\large
$

Note that there's a normalization $1/N$ step for the IDFT transformation.

**Where**:

-  $x[n] $: The discrete-time signal (time domain) with $n$ being the index of the sample.
-  $X[k]$: The discrete frequency components (frequency domain) at index $k$.
-  $N$: The total number of samples in the signal (window size).
-  $k$: The index of the frequency bin, ranging from 0 to $N-1$.
-  $n$: The index of the time sample, ranging from 0 to $N-1$#.
-  $e^{-j \frac{2\pi}{N} k n}$: The complex exponential representing the basis functions in the Fourier transform (used for decomposition into sinusoids), where $j$ is the imaginary unit.

## Algorithms & Optimizations

Naive DFT runs at $O(n^2)$, while FFT is much faster $O(n\log(n))$ with optimizations. We use the [Cooleyâ€“Tukey](https://en.wikipedia.org/wiki/Cooley%E3%80%93Tukey_FFT_algorithm) Radix-2 algorithm for power-of-two window sizes (e.g., 256, 512) and fall back to [Chirp Z Bluestein](https://en.wikipedia.org/wiki/Chirp_Z-transform) for arbitrary sizes.

Further optimizations like SIMD, memory management, and concurrency are possible but not implemented here.

pydelia implementation in `src/dsp/transforms.zig`

For a detailed explanation of FFT, check out [this resource](https://cp-algorithms.com/algebra/fft.html).

Below an example of the DFT algorithm (without the FFT optmizations and in python so we need a very small input)

```python
samples = [1.0, 0.75, 0.50,  0.25, 0, -0.25, -0.50, -0.75, -1.0]

def dft(samples, direction='forward'):
    output = []
    
    N = len(samples)
    sign = 1  if direction == 'inverse' else -1

    for k in range(N):
        ffted = complex(0, 0)
        for n in range(N): 
            angle = (2.0 * np.pi * k * n) / N
            exp = complex(np.cos(angle), sign * np.sin(angle))
            ffted += (exp * complex(samples[n], 0))

        # For inverse DFT, divide by N
        if direction == 'inverse':
            ffted /= N

        output.append(ffted)

    return output

dfted = dft(samples)
inv = dft(dfted, 'inverse')

# get only the real part after idft
inv_samples = []
for sample in inv:
    inv_samples.append(sample.real)

print(f"""
Input Samples: {samples}

The DFT Output is a list of complex numbers of the same size as the input:
{dfted}

Inverse DFT should reconstruct the original samples (subject to minor numeric inaccuracies):
{inv_samples}
""")
```

### Magnitude, Phase and the Frequency Bins

After transforming the signal into the frequency domain, we can compute the magnitude and phase of the complex FFT output. Using the FFT (or window) size, we can determine the frequency bins, allowing us to plot the transformed signal in the frequency domain.

**Magnitude**

$
|X[k]| = \sqrt(re_k^2 + im_k^2), \quad k = 0, 1, 2, \dots, N-1
$

**Phase**

$
\theta[k] = \text{atan2}(re_k, im_k), \quad k = 0, 1, 2, \dots, N-1
$

where $re$ and $im$ are the real and imaginary parts respectively.

**Frequency Bins**

$
f[k] = \frac{k \cdot f_s}{N}, \quad k = 0, 1, 2, \dots, N-1 
$

where $f_s$ is the sampling rate and $N$ the window size of the FFT.


**Negative Frequencies**

The FFT returns both positive and negative frequencies, with results spanning from 0 up to the sampling frequency $f_s$. However, we only care about frequencies up to the Nyquist frequency ($\frac{f_s}{2}$), as anything above this is a mirrored copy of the lower frequencies. These "negative frequencies" arise due to the symmetry of the Fourier transform for real-valued signals. To avoid this duplication, we discard the frequencies beyond the Nyquist limit.



```python
# our signal
sr = 44100 # hz
fft_size = 2048 # samples
freq = 1000 # hz
dur = fft_size / sr # to calculate the exact duration of the sine that we need for this example

# using our custom size wave as a helper
sine = pydelia.sine_wave(freq=freq, amp=0.5, sr=sr, dur=dur) 
t = np.linspace(0, dur, int(sr*dur), endpoint=False)

plt.plot(t, sine)
plt.title("1000Hz Sine Wave")
plt.ylabel("Amp")
plt.xlabel("Time [s]")
```

```python
# using our own FFT implementation here but could easily be using numpy as well

vec = pydelia.fft(sine)
N = len(vec)
mags = pydelia.magnitude(vec)
phase = pydelia.phase(vec)
bins = pydelia.fft_frequencies(N, sr)

inv = pydelia.ifft(vec)
inv_sine = []

# we only need the real part here
for sample in inv:
    inv_sine.append(sample.real)

fig, (ax1, ax2) = plt.subplots(1,2, figsize=(12, 5))

ax1.plot(bins[:N // 2], mags[:N // 2])
ax1.set_title ("FFT Magnitude of Sine Wave")
ax1.set_xlabel("Frequency (hz)")
ax1.set_ylabel("Magnitude")
ax1.set_xscale("log")
ax1.get_xaxis().set_major_formatter(mticker.ScalarFormatter()) 
ax1.grid(True)

ax2.plot(t, inv_sine)
ax2.set_title("Sine wave From IFFT")
ax2.set_xlabel("Time(s)")
ax2.set_ylabel("Amplitude")
```

```python
# and phase. Using numpy.wrap to keep the ranges between -pi and pi

phase  = np.unwrap(phase)

plt.plot(bins, phase[:N // 2])
plt.title("FFT Phase of Sine Wave")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Phase (radians)")
```

### Windowing Effect 


The window size in FFT affects the frequency resolution and the number of frequency bins, especially impacting the lower frequencies.

The frequency resolution is given by:

$
\Delta f = \frac{f_s}{N}
$

For example, with a sampling rate $f_s = 44.1 \text{kHz}$, using window sizes of $N = 1024$ and $N = 512$ will result in different frequency resolutions.

```python
fs = 44100
N1 = 1024
N2 = 512

print(f"""
 For fft size N = {N1} @ {fs}hz then {fs}/{N1} = {fs/N1}hz
 For fft size N = {N2} @ {fs}hz then {fs}/{N2} = {fs/N2}hz
 """
)
```

If we want to calculate the FFT of a smaller sample set while maintaining a larger FFT window for higher resolution, we can pad the samples with zeroes. This process doesn't add any new information to the signal, but it increases the number of FFT points, resulting in finer spacing between the frequency bins. The benefit is that we get a more detailed view of the signal's frequency content.

For example, we might have 64 samples but wish to achieve a frequency resolution of $\frac{f_s}{1024}$ instead of $\frac{f_s}{64}$.

```python
fs = 44100
nb_samples = 64
desired_window = 1024

print(f"""
      window/fft size = {nb_samples}, resolution: {fs/nb_samples} hz
      window/fft size = {desired_window}, resolution = {fs/desired_window} hz
""")
```

```python
samples = sine[:nb_samples]
# extend while padding with 0
padded_samples = sine[:nb_samples]
padded_samples.extend([0.0] * (desired_window - nb_samples))

print(f"""
 len(samples) = {len(samples)}
 len(padded_samples) = {len(padded_samples)}
""")
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(12, 3))

ax1.plot(t[:len(samples)], samples)
ax1.set_title("64 Samples")
ax1.set_xlabel("Time(s)")
ax1.set_ylabel("Amplitude")

ax2.plot(t[:len(padded_samples)], padded_samples)
ax2.set_title("1024 with 0 Padded")
ax2.set_xlabel("Time(s)")
ax2.set_ylabel("Amplitude")

```

**Padding will give us more frequency resolution, in other hands it will give us a finer spacing between frequency bins**

```python
vec = pydelia.fft(samples)
padded_vec = pydelia.fft(padded_samples)

mags = pydelia.magnitude(vec)
padded_mags = pydelia.magnitude(padded_vec)

bins = pydelia.fft_frequencies(len(vec), sr)
padded_bins = pydelia.fft_frequencies(len(padded_vec), sr)

fig, (ax1, ax2) = plt.subplots(1,2, figsize=(12, 5))

ax1.plot(bins, mags[:len(vec) // 2])
ax1.set_title ("64 Samples")
ax1.set_xlabel("Frequency (hz)")
ax1.set_ylabel("Magnitude")
ax1.grid(True)

ax2.plot(padded_bins, padded_mags[:len(padded_vec) // 2])
ax2.set_title ("1028 samples with padded 0")
ax2.set_xlabel("Frequency (hz)")
ax2.set_ylabel("Magnitude")
ax2.grid(True)
```

### Window Functions

Padding a signal to fit a larger FFT window can cause **spectral leakage** (see above), where energy from a frequency component spreads into nearby frequencies, particularly at higher frequencies. This leakage can be reduced by applying window functions like the **Hann** or **Blackman** window.

**Hann Window**:

$
w[k] = 0.5 \left( 1 - \cos \left( \frac{2\pi k}{N-1} \right) \right)
$

- Provides gradual tapering with moderate suppression of spectral leakage.
- Suitable for general-purpose applications where some leakage is acceptable.

**Blackman Window**:

$
w[k] = 0.42 - 0.5 \cdot \cos \left( \frac{2\pi k}{N-1} \right) + 0.08 \cdot \cos \left( \frac{4\pi k}{N-1} \right)
$

- Offers stronger suppression of spectral leakage.
- Ideal for applications requiring high dynamic range, where distant frequency components must be well suppressed.
- This comes with a slight reduction in frequency resolution.

**Normalization**
In audio applications, it is common to normalize the window function to ensure **preservation of total energy** in the signal. This is done by dividing the windowed signal by the sum of the window coefficients:

$
x_w[k] = \frac{w(k) \cdot x(k)}{\sum_k w(k)}
$

Here, $ w(k) $ is the window coefficient at sample $( k )$, and $ \sum_k w(k) $ ensures that the energy remains consistent after windowing.

Below, we apply the window functions to the 64 samples. Due to the small sample size, the resulting plots may not show significant differences.

```python

def plot(s, name):
    hann =   pydelia.hanning(s)
    blackman = pydelia.blackman(s)
    
    # Both signals have the same size
    N = len(hann)

    hann_vec = pydelia.fft(hann)
    blackman_vec = pydelia.fft(blackman)
    
    hann_mags = pydelia.magnitude(hann_vec)
    blackman_vec = pydelia.magnitude(blackman_vec)
    
    bins = pydelia.fft_frequencies(N, sr)
    
    fig, (ax1, ax2) = plt.subplots(1,2, figsize=(12, 5))
    
    ax1.plot(bins, hann_mags[:N // 2] )
    ax1.set_title (f"{name} :Hanning Window")
    ax1.set_xlabel("Frequency (hz)")
    ax1.set_ylabel("Magnitude")
    ax1.grid(True)
    
    ax2.plot(bins, hann_mags[:N // 2] )
    ax2.set_title (f"{name}: Blackman Window")
    ax2.set_xlabel("Frequency (hz)")
    ax2.set_ylabel("Magnitude")
    ax2.grid(True)

plot(samples, "Samples")
plot(padded_samples, "Padded Samples")


```

### Spectrogram: Time-frequency Representation

A spectrogram is a time-frequency representation that provides an estimate of the short-time, time-localized frequency content of a signal.

To compute the spectrogram, the signal is divided into segments of length \( N \), each of which is multiplied by a window function (to reduce spectral leakage), and then a Fast Fourier Transform (FFT) is applied to each segment. To improve the time-localization of the short-time spectra, the windowed segments are often overlapped.

Visually, a typical spectrogram displays time along the horizontal (**$X$**) axis and frequency along the vertical (**$Y$**) axis. Each vertical slice represents the magnitude of the signal's frequency content at a specific time, with the magnitudes displayed using a grayscale or heatmap color scale. The higher the magnitude, the brighter or more intense the color at that point in the spectrogram.

Let's import an actual audio file to have a more interesting spectogram.


```python
samples, sr = librosa.load('audio_examples/voice.wav')
#one second
audio = samples[int(1.0 * sr): (int(2.0 *sr))]

print(f"""
Random 10 samples
     {audio[100:110]}
""")

plt.figure(figsize=(10, 4))
plt.plot(audio)
plt.title('Audio 1 second')
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

First, let's use our FFT implementation to get the frequency doment of of a single segment of our audio.

```python
N = 1024

# we see above that around 3000, 4000 samples, we have some audio signal
segment = audio.tolist()[3500:3500 + N]

vec = pydelia.fft(segment)
mags = pydelia.magnitude(vec)
bins = pydelia.fft_frequencies(N, sr)
plt.figure(figsize=(10, 5))
plt.plot(bins, mags[: N // 2])
plt.title("Single Segment Frequency Domain Plot")
plt.xlabel("Frequency (hz)")
plt.ylabel("Magnitude")
plt.show()

```

```python
def spectogram(samples, sr, window_size, hop_perc=0.5):
    n_samples = len(samples)
    hop_size = int(window_size * hop_perc) # so for 0.5 hop percent and 1024 window hop_size 1024 * 0.5 = 512

    # the number of windows our audio has
    # so for a second of audio @ 44K1 and window_size = 1024 ->  ((44100 -  1024) /  512) + 1 = 85.1328125 (round to 85)
    n_windows = (n_samples - window_size) // hop_size + 1
    
    # Create the spectrogram array (N/2 for frequencies, n_windows for time)
    spectogram = np.zeros((window_size // 2, n_windows)) 

    for window_index in range(n_windows):

        start = window_index * hop_size
        segment = samples[start:start + window_size]
        windowed_segment = pydelia.hanning(segment)

        fft_result = pydelia.fft(windowed_segment)
        mags = pydelia.magnitude(fft_result)
        db_mags = pydelia.decibels_from_magnitude(mags)
        spectogram[:, window_index] = np.array(db_mags[: window_size // 2])

    bins = pydelia.fft_frequencies(window_size, sr)
    plt.figure(figsize=(10,6))
    
    plt.imshow(spectogram,
               origin='lower',
               aspect='auto',
               # sets x axis to be between 0 and n_samples/sr seconds
               # and y axis to be between our frequency bins
               extent=[0, n_samples / sr, bins[0], bins[len(bins)-1]],
               cmap='viridis'
              )

    plt.colorbar(label="Magnitude (DB)")
    plt.title("Spectogram")
    plt.xlabel("Time (s)")
    plt.ylabel("Frequency (Hz)")

    plt.show()
    print(spectogram.shape)

spectogram(audio.tolist(), sr, N, 0.3)
```

delia is also exposes to python a short fourier transform for testing purposes, which basically takes the spectogram as above

```python
delia_matrix = pydelia.stft(audio.tolist(), 2048, 2048 // 4)

def plot_delia(matrix, sr):
     matrix = np.array(matrix)
     rows, cols = matrix.shape
     spectro = np.zeros((rows, cols))

     for c in range(cols):
        row = matrix[:, c]
        
        row = pydelia.magnitude(row.tolist())
        row = pydelia.decibels_from_magnitude(row, np.max(row).item())
        spectro[:, c] = row 

     bins = pydelia.fft_frequencies(rows, sr)

     plt.figure(figsize=(10,6))
     plt.imshow(spectro,
             origin='lower',
             interpolation='bilinear',
             aspect='auto',
             extent=[0, len(audio) / sr, bins[0], bins[len(bins)-1]],
             cmap='magma'
               )

     plt.colorbar(label="Magnitude (DB)")
     plt.title("Spectogram")
     plt.xlabel("Time (s)")
     plt.ylabel("Frequency (Hz)")
     plt.show()
    
    
plot_delia(delia_matrix, sr)
```

## **Convolution**

Convolution is a fundamental operation in signal processing, analogous to addition or multiplication, but it combines two signals in a specific way. It can be performed both in the time domain, often using a **finite impulse response (FIR)** filter, and in the frequency domain by multiplying the spectra of the signals using the **Fast Fourier Transform (FFT)**.

In the frequency domain, convolution is represented as:

$$
\large
C(f) = A(f) \cdot B(f)
\large
$$

Where:
-  $A(f) $ and  $B(f) $ are the Fourier transforms of the signals.
-  $C(f) $ is the product of the two spectra, representing the convolved signal.

### **Time Domain Convolution**

In the time domain, convolution involves one signal being **filtered** by another, often referred to as the **kernel**. The result is calculated by sliding the kernel across the signal and summing the element-wise products. For two signals  $x(n) $ and  $h(n) $, the convolution is defined as:

$$
\large
y(n) = \sum_{k=0}^{M-1} x(n - k) h(k)
\large
$$

Where:
-  $x(n) $ is the input signal,
-  $h(k) $ is the kernel (or filter),
-  $y(n) $ is the resulting signal,
-  $M $ is the length of the kernel.


For example, convolving a signal with a smoothing kernel can filter high frequencies or noise, acting as a lowpass filter.

### **Interpreting Time Domain Convolution**

The time domain interpretation of convolution can be understood by looking at how the **kernel** affects the input signal. Hereâ€™s a clearer breakdown:

1. **Single pulse as the kernel**: If the kernel  $h(n) $ is a single pulse (like a Dirac delta function), convolution shifts the input signal by a certain amount. For example, if  $h(n) $ is a pulse at time  $k_0 $, the result will be a **shifted copy** of  $x(n) $ by  $k_0 $. This is equivalent to delaying the input signal.

2. **Multiple pulses as the kernel**: If  $h(n) $ contains **multiple pulses**, the result will be a copy of  $x(n) $ at each pulse position. This can create rhythmic or reverberant structures, depending on the spacing between the pulses. Essentially, each pulse in  $h(n) $ creates a shifted version of  $x(n) $, resulting in a rhythmic effect.

3. **Pulse-like kernel**: If  $h(n) $ is more **pulse-like** (narrow peaks), it acts like a filter, shaping the input signal  $x(n) $. In this case, the kernel  $h(n) $ can be viewed as the **impulse response** of a system. For example, convolving  $x(n) $ with the sound of a bell would give a result that simulates the effect of hitting a bell. This allows us to recreate resonances without explicitly measuring them.

4. **Complex signals**: When both  $x(n) $ and  $h(n) $ are **complex signals** with rich time and frequency content, the resulting signal tends to be **blurred**. The original character of the sounds may be lost due to overlapping frequency components. If both signals have **long durations** and strong pitches, the result will contain both pitches, along with interactions between their spectra, leading to more complex sounds.


Below an example of delaying the signal

```python
fs=44100
fc= 200
dur= 1 / fc * 4 # 4 cycles

x = pydelia.sine_wave(freq= fc, sr=fs, amp=0.8, dur=dur)
n = len(x)

print(n)

# impulse 
h = np.zeros(len(x))
h[100] = 1 # delayed

y = signal.convolve(np.array(x), h, mode='full')
plt.figure(figsize=(10, 5))
plt.subplot(3, 1, 1)
plt.title("Input Signal(x)")
plt.plot(np.arange(n), np.array(x))
plt.grid(True)
plt.xlim(0, n)

plt.subplot(3, 1, 2)
plt.title("Kernel (h) Dirac Delta")
plt.plot(np.arange(n), h)
plt.grid(True)
plt.xlim(0, n)

plt.subplot(3, 1, 3)
plt.title("Dirac Output Signal (y = x * h)")
plt.plot(np.arange(len(y)), y)
plt.grid(True)
plt.xlim(0, n)

plt.tight_layout()
plt.show()
```
